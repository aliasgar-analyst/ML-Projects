# Part 1

```python
import pandas as pd
import numpy as np

# 1) Load the given dataset
df = pd.read_csv(r"D:\Ali\Project Portfolio\ML Projects\Linear Reg\ML Linear Regression Home Price Prediction & Analysis Project.txt")
df
```

```python
# 2) Print all the column names
print("Column Names:", df.columns.tolist())
```

```python
# 3) Describe the data
print("\nData Description:\n", df.describe())
```

```python
# 4) Drop address, date, postcode, YearBuilt, lattitude, longtitude columns
df = df.drop(['Address', 'Date', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude'], axis=1)
```

```python
# 5) Find the count of null value in each column
print("\nNull Values Count:\n", df.isnull().sum())
```

```python
# 6) Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0
cols_to_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_zero] = df[cols_to_zero].fillna(0)
```

```python
# 7) Fill Null value of land size and bidding area (BuildingArea) columns with Mean
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
# 8) Drop rows where Price (Target) is null before ML
df.dropna(subset=['Price'], inplace=True)
```

```python
# 9) Find the unique value in method column 
print("\nUnique values in Method:", df['Method'].unique())
```

```python
# 10) Create a dummy data for categorical data 
df = pd.get_dummies(df, drop_first=True)
```

# Part 2

```python
# 1) Create the target data and feature data where target data is price 
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
# Splitting for validation
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
# 2) Create a linear regression model for Target and feature data 
from sklearn.linear_model import LinearRegression, Lasso, Ridge

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# 3) Check if the model is overfitting or underfitting or it is accurate 
train_score = reg.score(X_train, y_train)*100
test_score = reg.score(X_test, y_test)*100

print(f"\nTraining Score: {train_score}")
print(f"Testing Score: {test_score}")
```

```python
# 5) Extract slope and intercept value from the model 
print("\nSlope (Coefficients):", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
# Predictions for metrics
y_pred = reg.predict(X_test)
```

```python
# 6) Display Mean Squared Error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error (MSE):", mse)
```

```python
# 7) Display Mean Absolute Error 
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
# 8) Display Root mean Squared error 
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# 9) Display R2 score 
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```
