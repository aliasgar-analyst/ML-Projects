# Part 2

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 1) Perform encoding in columns Gender, Customer_Type, Type_of_Travel, and Class 
le = LabelEncoder()
encode_cols = ['Gender', 'Customer_Type', 'Type_of_Travel', 'Class']
for col in encode_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# 2) Drop the column id and Unnamed:_0.1
df = df.drop(['id', 'Unnamed:_0.1'], axis=1)
```

```python
# 3) Create the features and target Data 
X = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
# 4) Perform scaling on features data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# 5) Split the data in training and testing sets 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# 6) Fit the decision tree model with various parameters 
# You can experiment with 'gini' and max_depth
dt_model = DecisionTreeClassifier(criterion='gini', max_depth=10, random_state=42)
dt_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix, classification_report)

# 7) Create a function to display metrics
def display_metrics(y_true, y_pred):
    print("Accuracy Score:", accuracy_score(y_true, y_pred))
    print("Precision Score:", precision_score(y_true, y_pred))
    print("Recall Score:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("\nConfusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("\nClassification Report:\n", classification_report(y_true, y_pred))
```

```python
# Generate predictions and display metrics
y_pred = dt_model.predict(X_test)
display_metrics(y_test, y_pred)
```
